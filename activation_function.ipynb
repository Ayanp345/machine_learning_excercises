```python
import numpy as np
import matplotlib.pyplot as plt

# Define input range
x = np.linspace(-10, 10, 500)

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

# Create plots
plt.figure(figsize=(12,4))

plt.subplot(1, 3, 1)
plt.plot(x, sigmoid(x), label="Sigmoid", color='blue')
plt.title("Sigmoid")
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(x, tanh(x), label="Tanh", color='green')
plt.title("Tanh")
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(x, relu(x), label="ReLU", color='red')
plt.title("ReLU")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Activation functions
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def swish(x, beta=1.0):
    return x * (1 / (1 + np.exp(-beta * x)))

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=-1, keepdims=True)

# Generate input data
x = np.linspace(-5, 5, 1000)  # Range for ELU, Swish, Leaky ReLU
x_softmax = np.linspace(-5, 5, 100)  # Smaller range for Softmax demo
# For Softmax, create a 2D array simulating 3 classes with varying inputs
softmax_input = np.vstack([x_softmax, x_softmax * 0.5, x_softmax * 0.2]).T

# Compute outputs
elu_output = elu(x)
swish_output = swish(x)
leaky_relu_output = leaky_relu(x)
softmax_output = softmax(softmax_input)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot ELU
axs[0, 0].plot(x, elu_output, label='ELU', color='blue')
axs[0, 0].set_title('ELU Activation')
axs[0, 0].set_xlabel('Input')
axs[0, 0].set_ylabel('Output')
axs[0, 0].grid(True)
axs[0, 0].legend()
axs[0, 0].axhline(0, color='black', linewidth=0.5, linestyle='--')
axs[0, 0].axvline(0, color='black', linewidth=0.5, linestyle='--')

# Plot Swish
axs[0, 1].plot(x, swish_output, label='Swish', color='green')
axs[0, 1].set_title('Swish Activation')
axs[0, 1].set_xlabel('Input')
axs[0, 1].set_ylabel('Output')
axs[0, 1].grid(True)
axs[0, 1].legend()
axs[0, 1].axhline(0, color='black', linewidth=0.5, linestyle='--')
axs[0, 1].axvline(0, color='black', linewidth=0.5, linestyle='--')

# Plot Leaky ReLU
axs[1, 0].plot(x, leaky_relu_output, label='Leaky ReLU', color='red')
axs[1, 0].set_title('Leaky ReLU Activation')
axs[1, 0].set_xlabel('Input')
axs[1, 0].set_ylabel('Output')
axs[1, 0].grid(True)
axs[1, 0].legend()
axs[1, 0].axhline(0, color='black', linewidth=0.5, linestyle='--')
axs[1, 0].axvline(0, color='black', linewidth=0.5, linestyle='--')

# Plot Softmax (showing probabilities for 3 classes)
axs[1, 1].plot(x_softmax, softmax_output[:, 0], label='Class 1', color='purple')
axs[1, 1].plot(x_softmax, softmax_output[:, 1], label='Class 2', color='orange')
axs[1, 1].plot(x_softmax, softmax_output[:, 2], label='Class 3', color='cyan')
axs[1, 1].set_title('Softmax Activation (3 Classes)')
axs[1, 1].set_xlabel('Input (Class 1)')
axs[1, 1].set_ylabel('Probability')
axs[1, 1].grid(True)
axs[1, 1].legend()
axs[1, 1].set_ylim(0, 1)  # Probabilities range from 0 to 1

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plot
plt.show()
```
